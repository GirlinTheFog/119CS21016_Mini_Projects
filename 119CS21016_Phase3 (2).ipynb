# 4.Visualize above with Bar, Box, FacetGrid, Heatmap, Crosstab, Line plot


```python
# import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Replace these with the names of the numerical and categorical variables you want to analyze
numerical_variable = 'AGE'
categorical_variable = 'DAY_OF_WEEK'

# Bar plot for a categorical variable
plt.figure(figsize=(10, 6))
sns.countplot(x=categorical_variable, data=df, palette="Set1")
plt.title(f'Bar Plot: {categorical_variable}')
plt.xlabel(categorical_variable)
plt.ylabel('Count')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()


```

```python
# Box plot for numerical vs. categorical variable
plt.figure(figsize=(10, 6))
sns.boxplot(x=categorical_variable, y=numerical_variable, data=df, palette="Set2")
plt.title(f'Box Plot: {numerical_variable} vs. {categorical_variable}')
plt.xlabel(categorical_variable)
plt.ylabel(numerical_variable)
plt.xticks(rotation=45)
plt.show()
```

```python
# FacetGrid for visualizing the relationship between two categorical variables
g = sns.FacetGrid(df, col=categorical_variable, height=5)
g.map(sns.countplot, numerical_variable, palette="Set3")
g.set_axis_labels(numerical_variable, "Count")
plt.subplots_adjust(top=0.8)
g.fig.suptitle(f'FacetGrid: {numerical_variable} by {categorical_variable}')
plt.show()

```

```python
# Crosstab for a cross-tabulation of two categorical variables
crosstab = pd.crosstab(df[categorical_variable], df['Sex'])
crosstab.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title(f'Cross-tabulation: {categorical_variable} vs. Sex')
plt.xlabel(categorical_variable)
plt.ylabel('Count')
plt.show()

```
